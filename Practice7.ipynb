## Постановка задачи

Необходимо решить методом динамического программирования одну из следующих задач:
1)	Требуется подсчитать количество последовательностей длины $N$, состоящих из 0 и 1, в которых никакие две единицы не стоят рядом. На вход подается целое число N. На выходе: список найденных последовательностей.
2)	Задача о размене монет. В торговом автомате имеется определенный набор монет `COINS`, количество монет каждого номинала не ограничено. Определить, каким минимальным количеством монет можно выдать сдачу N. На входе: массив номиналов монет `COINS = [coin1, coin2 ,… , coinM]` и сумма $N$, которую нужно выдать. На выходе: набор монет `{coin1: k1, coin2: k2, …, coinM: kM}`, который в сумме дает $N$.
3)	Задача о дискретном рюкзаке. Имеется ряд ценностей, для которых заданы масса и общая стоимость; максимальная вместимость рюкзака ограничена (по массе); необходимо определить, какие ценностей нужно взять, чтобы их стоимость была максимальная, а масса не превышала вместимость рюкзака; каждый вид ценности можно взять только целиком. На входе: N – максимально допустимая масса ценностей, которые может выдержать рюкзак и набор ценностей, описываемых парой значений (масса, стоимость): `THINGS = [(масса1; стоимость1), (масса2; стоимость2), …, (масса_n; стоимость_n)]`. На выходе: список номеров ценностей из массива `THINGS`, которые имели бы максимальную стоимость и их масса не превышала бы значение N.

## Задача 1

```python
def binaries_without_consecutive_ones(n: int) -> list:
    if n <= 0:
        raise ValueError

    zero_ended = ['0']
    one_ended = ['1']

    for i in range(1, n):
        zero_ended_next = [item + '0' for item in zero_ended + one_ended]
        one_ended_next = [item + '1' for item in zero_ended]

        one_ended = one_ended_next
        zero_ended = zero_ended_next

    return zero_ended + one_ended
```

```python
for i in range(1, 8):
    print(list(binaries_without_consecutive_ones(i)))
```
